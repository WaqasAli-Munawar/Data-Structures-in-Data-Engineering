In a stack, the last plate to go in will be the first plate to be washed. For this reason, in programming, stacks are called a `last-in, first-out (LIFO) data structure`.

Just like the queue we've implemented in the previous file, our stack implementation is going to be based on the linked list.

```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class LinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
        
    def append(self, data):
        new_node = Node(data)
        if self.length == 0:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1
        
    def __iter__(self):
        self._iter_node = self.head
        return self 
    
    def __next__(self):
        if self._iter_node is None:
            raise StopIteration
        ret = self._iter_node.data
        self._iter_node = self._iter_node.next
        return ret
    
    def prepend(self, data):
        new_node = Node(data)
        if self.length == 0:
            self.head = self.tail = new_node
        else:
            self.head.prev = new_node
            new_node.next = self.head
            self.head = new_node
        self.length += 1
        
    def __len__(self):
        return self.length
    
    def __str__(self):
        return str([value for value in self])
```

```python
class Stack(LinkedList):
    
    def push(self, data):
        self.append(data)

    # Add peek() method here
    def peek(self):
        return self.tail.data
        
stack = Stack()
for i in [1, 2, 3]:
    stack.push(i)
    
print(stack.peek())
```

To complete the main functionality of our stack, we need to implement a method that removes and retrieves the top of the stack. This method is commonly known as **pop**.

The top of the stack is represented by the tail node. This is the same node was used to represent the front of the queue. Therefore, the implementation of the push method is the same as the implementation of the `dequeue()` method we did in the previous file!

In any of these cases, we need to update the `length` attribute of the stack by decrementing it by one.

```python
class Stack(LinkedList):
    
    def push(self, data):
        self.append(data)

    def peek(self):
        return self.tail.data

    # Add pop() method here
    def pop(self):
        ret = self.tail.data
        if self.length == 1:
            self.tail = self.head = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        self.length -= 1
        return ret

stack = Stack()
for i in [1, 2, 3]:
    stack.push(i)
top = stack.pop()
print(stack.peek())
```

Our stack implementation is now complete. Stacks have a lot of applications. For instance, stacks are used to manage function call arguments in most programming languages. Another application of stacks is in evaluating arithmetic expressions in a computer program.

The simulation will build two new columns in the `processes` DataFrame. One named `Start` which will provide the start time of each process. The second, named `End` will store the time at which the process finishes its execution.

The simulation process is the same as in the previous file. Let's recap how it worked. We'll simulate time by incrementing the `cur_time` variable. At each time, we need to do three things:



Since we've already implemented it once in the previous file, we provide with higher level template and challenge to fill in the gaps.

The columns our algorithms needs to calculate are the last two: `Start` and `End`.

**Task**



**Answer**

```python
cur_time = 0
num_processes_done = 0
wait_stack = Stack()
cur_pid = None

while num_processes_done < processes.shape[0]:
    # Step 1 goes here
    if cur_pid is not None:
        if processes.loc[cur_pid, "Start"] + processes.loc[cur_pid, "Duration"] == cur_time:
            processes.loc[cur_pid, "End"] = cur_time
            cur_pid = None
            num_processes_done += 1
    # Step 2 goes here
    ready_processes = processes[processes["Arrival"] == cur_time]
    for pid, _ in ready_processes.iterrows():
        wait_stack.push(pid)
    # Step 3 goes here
    if cur_pid is None and len(wait_stack) > 0:
        cur_pid = wait_stack.pop()
        processes.loc[cur_pid, "Start"] = cur_time 
    cur_time += 1
    
processes.head()
```

Above, we implemented the FCFS algorithm and calculated the start and end times of each process in the `processes` DataFrame.
