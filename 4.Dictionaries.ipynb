We are provided with some code that we need to complete. The idea is to insert all values in the `keys` list into a dictionary `d` and measure the execution time of each insertion.



**Answer**

```python
import matplotlib.pyplot as plt

def plot_times(times):
    plt.plot(times)
    plt.show()

import time
import random
random.seed(0)

times = []
number_of_entries = 1000
keys = range(number_of_entries)

d = dict()

for key in keys:
    start = time.time()
    d[key] = key
    end = time.time()
    times.append(end - start)
    
plot_times(times)
```

Above we learned that the time complexity of dictionary operations depends on the load factor. To reduce the load factor, the number of buckets is increased as we add more entries to the dictionary.
